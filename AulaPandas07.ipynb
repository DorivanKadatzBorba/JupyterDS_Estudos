## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados.head(50)
```

```python
# Valor medio do aluguel geral
dados['Valor'].mean()
```

```python
# A atividade consiste em saber o preço médio de cada bairro
```

```python
# Obtendo os bairros sem duplicidades
dados['Bairro'].drop_duplicates()
```

```python
# Variável que recebe o data frame 'dados' agrupado por bairros
grupo_bairro = dados.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
# A váriabel está guardando um objeto do tipo DataFrameGoupBy
grupo_bairro
```

```python
''' Ao fazer um goupby na base de dados, a variável grupo_bairro recebeu um data frame para cada bairro diferente
    sendo possível trabalhar com todos os elementos que estejam na tabela desde que seja relacionado com o bairro'''
```

```python
# Indica o nome do bairro e qual indice onde esse bairro foi encontrado no data frame
grupo_bairro.groups
```

```python
# print de todos os bairros diferentes
for bairro, dados in grupo_bairro:
    print(bairro)
```

```python
# Print de todas as tabelas agrupadas por bairro
for bairro, dados in grupo_bairro:
    print(dados)
```

```python
# A variável grupo bairro
for bairro, dados in grupo_bairro:
    print(type(dados))
```

```python
for bairro, dados in grupo_bairro:
    print(f'{bairro} --> {dados.Valor.mean()}')
```

```python
valor_medio = pd.DataFrame(grupo_bairro[['Valor', 'Condominio']].mean().round(2))
```
