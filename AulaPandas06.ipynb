## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep =';')
dados.head(10)
```

```python
# Nova váriavel contendo a soma de 'Valor' + 'Condominio' + 'IPTU'
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
```

```python
# Valor do metro quadrado
dados['Valor m2'] = dados['Valor'] / dados['Area']
dados.head(10)
```

```python
# Arredondadndo o valor da variavel 'Valor m2' em duas casas decimais
dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head(10)
```

```python
# Valor bruto por metro quadrado
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(10)
```

```python
casas = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
'''Apply' permite aplicar uma função para cada linha do data frame.

 Criará uma nova variaval com nome 'Tipo agregado' que receberá os valor 'Casa' se o valor na variável 'Tipo'
 estiver na lista "casas" criada anteriormente, caso contratio receberá o valor Apartamento''' 
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casas else 'Apartamento')
dados.head(10)
```

## Excluindo variáves de um data frame

```python
# Data Frame apendas com as novas variaveis criadas com o data frame anterior
# Não esquecer do colchete duplo
dados_aux = pd.DataFrame(dados[['Valor Bruto', 'Valor m2', 'Valor Bruto m2', 'Tipo Agregado']])
dados_aux.head(10)
```

```python
# Removendo variaveis utilizando 'del'
del dados_aux['Valor Bruto']
dados_aux.head(10)
```

```python
# Utilizando 'pop' para remoção irá retornar a variavel que foi removida
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux.head(10)
```

```python
# Utilizando drop é possível excluir mais de uma variável
#A atividade pedia para remover as variaveis do dataframe principal 'dados'
# OBS: utilizar 'axis = 1' para que o codigo trabelhe nas colunas, caso contraio irá deletar a linha caso exista o valor citado.
# Inplace True para que a auteração seja feita diretamente
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
dados.head(10)
```

## Exportando a base de dados

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
