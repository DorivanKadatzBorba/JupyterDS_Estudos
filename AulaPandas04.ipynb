## (1) Seleção de Imoveis do tipo Apartamento

```python
import pandas as pd
```

```python
# Importdando dados
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head(10)
```

```python
# Verificando se 'Apartamento' está contido na base de dados
selecao = dados['Tipo'].isin(['Apartamento'])
selecao.head(10)
```

```python
apartamento = dados[selecao]
apartamento.head(10)
```

```python
# Verificando se na nova seleção existe apenas o tipo Apartamento
list(apartamento['Tipo'].drop_duplicates())
```

```python
# Quantidade de tuplas total de origem
dados.shape[0]
```

```python
# Quantidade de tuplas apenas com tipo apartamento
apartamento.shape[0]
```

```python
# Organizando o index
apartamento.index = range(apartamento.shape[0])
apartamento
```

# Exportando a base de dados

```python
# Base de dados exportada ignorando o index para não duplicar
apartamento.to_csv('dados/aluguel_apartamento.csv', sep = ';', index = False)
```

```python
# Verificando se está tudo ok com a exportação
dados = pd.read_csv('dados/aluguel_apartamento.csv', sep = ';')
dados
```

# (2) Seleção dos imóveis dos tipos 'casa', 'casa de condominio' e 'casa de vila'

```python
# Importando base de dados
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head(10)
```

```python
# Verificando se contem a lista de parametros dentro da coluna Tipo na base de dados
selecao2 = dados['Tipo'].isin(['Casa', 'Casa de Condomínio', 'Casa de Vila'])
selecao2.head(20)
```

```python
# Atribuindo os valores True na nova base de dados 
dados_selecao2 = dados[selecao2]
dados_selecao2
```

```python
# Verificando se a nova base de dados realmente contem apenas os tipos desejados
list(dados_selecao2['Tipo'].drop_duplicates())
```

```python
# Organizando o index
dados_selecao2.index = range(dados_selecao2.shape[0])
dados_selecao2
```

# Exportando a base de dados

```python
dados_selecao2.to_csv('dados/aluguel_selecao_casas.csv', sep = ';', index = False)
```
